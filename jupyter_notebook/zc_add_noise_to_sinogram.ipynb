```python
import os
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
sinogram_gt = np.load(os.path.join('movie_0_100_sinogram.npy'),allow_pickle = True)
print(sinogram_gt.shape)
plt.imshow(sinogram_gt,cmap="gray")
```

```python
# add Poisson noise
N0 = 200
counts = np.exp(sinogram_gt) * N0
counts_noise = np.random.poisson(counts)
sinogram_noise = np.log(counts_noise/N0)
plt.imshow(sinogram_noise,cmap="gray")
```

```python
sinogram_binary = sinogram_noise > 0.05
plt.imshow(sinogram_binary,cmap="gray")
```

```python
# calculate the CNR in sinogram

signal = sinogram_noise[sinogram_noise > 0.05]
background = sinogram_noise[sinogram_noise< 0.03]

CNR = (np.mean(signal)-np.mean(background))/np.std(background)
print(CNR)
```
